```python
## Count each event

## A list of events
events = ["train:delays", "train:on_time", "hotel:check_in", "train:delays",
           "snack:coffee", "train:on_time", "hotel:check_out", 
           "snack:tea", "train:delays"]
def count_events(entries):
    counts = {} # set up an empty dictionary to hold counts
    for entry in entries:
        category = entry.split(":")[0] # create sublists by splitting on ":" and selcting the first element in each list
        counts[category] = counts.get(category, 0) + 1 # accumulate counts for each category
    return counts

category_counts = count_events(events)

print(category_counts)

more_events = ["walk:city", "walk:park", "photo:landscape", 
               "photo:cafe", "train:on_time"]
print(count_events(more_events))
```
