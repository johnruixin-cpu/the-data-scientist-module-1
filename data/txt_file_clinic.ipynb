```python
from pathlib import Path
```

```python
data_folder = Path(r"C:\Users\johnr\Projects\the-data-scientist\week1_python_foundations\data")

message_path = data_folder / "messages.txt"

text = message_path.read_text(encoding="utf-8")

print(text)
```

```python
## read and inspect the messages
lines = text.splitlines() ## split the text into lines
print(f"Number of lines: {len(lines)}")
print("First line:", lines[0])
```

```python
def count_messages(lines):
    return len(lines)

def count_by_sender(lines):
    counts = {}
    for line in lines:
        sender = line.split(",", 1)[0] ## split the line by comma and take the second element (sender)
        counts[sender] = counts.get(sender, 0) + 1
    return counts
```

```python
total = count_messages(lines)

by_sender = count_by_sender(lines)
print(by_sender)
print(f"Total messages: {total}")
print("Messages by sender:")

for name, count in by_sender.items():
    print(f"{name}: {count}")
```

```python
summary_lines = []
summary_lines.append(f"Total messages: {total}")
summary_lines.append("")
summary_lines.append("Messages by sender:")
for name, count in sorted(by_sender.items(), key = lambda item: item[1], reverse=True):
    summary_lines.append(f" - {name}: {count}")

summary_text = "\n".join(summary_lines)

with open(data_folder / "message_summary.txt", "w", encoding="utf-8") as f:
    summary_file = f.write(summary_text)

with open(data_folder / "message_summary.txt", "r", encoding="utf-8") as f:
    final_summary = f.read()
print(f"Wrote summary to {final_summary}")
```
